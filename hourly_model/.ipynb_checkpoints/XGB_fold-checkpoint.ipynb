# Define functions

```python
def leave_one_out_xgb(X, y, fold, n_estimators, lr, depth, n_jobs, gamma, min_cw, subsample, colsample):
    uid = np.unique(fold)
    xgb_pred = np.zeros(X.shape[0])
    xgb_valid_loss = np.zeros(len(uid))
    xgb_train_loss = np.zeros(len(uid))
    for i in uid:
        x_valid = X[fold==i]
        x_train = X[fold!=i]
        y_valid = y[fold==i]
        y_train = y[fold!=i]
        model = XGBRegressor(n_estimators=n_estimators, learning_rate=lr, 
                           max_depth = depth, n_jobs = n_jobs, 
                           gamma = gamma, min_child_weight = min_cw,
                           subsample = subsample, colsample_bytree = colsample, random_state=1234)
        model.fit(x_train, y_train)

        pred = model.predict(x_valid)
        xgb_pred[fold==i] = pred
        xgb_valid_loss[uid==i] = mean_squared_error(y_valid, pred)
        xgb_train_loss[uid==i] = mean_squared_error(y_train, model.predict(x_train))
    return {'pred':xgb_pred, 'valid_loss':xgb_valid_loss, 'train_loss':xgb_train_loss}


def xgboostcv(n, p, depth, g, min_cw, subsample, colsample):
    xgb_loo = leave_one_out_xgb(X_train, y_train, fold_train, 
                                int(n)*100, 10**p, int(depth), 96, 
                                10**g, min_cw, subsample, colsample)
    MSE = mean_squared_error(y_train, xgb_loo['pred'])
    return -MSE

def performance_table(model, y, fold):
    uid = np.unique(fold)
    pred = np.round(model['pred'])
    pred[pred<0] = 0
    y = y.reshape(-1)
    xgb_valid_mse = np.zeros(len(uid))
    xgb_valid_mae = np.zeros(len(uid))
    xgb_valid_r2 = np.zeros(len(uid))
    for i in uid:
        pred_i = pred[fold==i]
        y_i = y[fold==i]
        xgb_valid_mse[uid==i] = mean_squared_error(y_i, pred_i)
        xgb_valid_mae[uid==i] = np.abs(pred_i-y_i).mean()
        xgb_valid_r2[uid==i] = r2_score(y_i, pred_i)
    
    results = pd.DataFrame(0, index=uid, 
                           columns=['xgb_valid_mse', 'xgb_valid_mae', 'xgb_valid_r2', 
                                    'xgb_valid_loss', 'xgb_train_loss'])
    results['xgb_valid_mse'] = xgb_valid_mse
    results['xgb_valid_mae'] = xgb_valid_mae
    results['xgb_valid_r2'] = xgb_valid_r2
    results['xgb_valid_loss'] = model['valid_loss']
    results['xgb_train_loss'] = model['train_loss']
    print(results)

def overall_performance(y, y_pred):
    overall_MSE = mean_squared_error(y, y_pred)
    overall_MAE = (np.abs(y_pred-y)).mean()
    overall_RMSE = np.sqrt(np.square(y_pred-y).mean())
    overall_R2 = r2_score(y, y_pred)
    print("XGB overall MSE: %0.4f" %overall_MSE)
    print("XGB overall MAE: %0.4f" %overall_MAE)
    print("XGB overall RMSE: %0.4f" %overall_RMSE)
    print("XGB overall R^2: %0.4f" %overall_R2)    

def plot_importance(model, columns):
    importances = pd.Series(model.feature_importances_, index = columns).sort_values(ascending=False)
    n = len(columns)
    plt.figure(figsize=(4,6))
    plt.barh(np.arange(n)+0.5, importances)
    plt.yticks(np.arange(0.5,n+0.5), importances.index)
    plt.ylim([0,n])
    plt.gca().invert_yaxis()
    
def save_model(clf, path):
    clf_pkl_fl = open(path, 'wb')
    pickle.dump(clf, clf_pkl_fl)
    clf_pkl_fl.close()
    
def load_model(path):
    f = open(path, 'rb')
    clf = pickle.load(f)
    f.close()
    return(clf)
```

```python
xgboostBO = BayesianOptimization(xgboostcv,
                                 {'n': (1, 10),
                                  'p': (-4, 0),
                                  'depth': (2, 10),
                                  'g': (-3, 0),
                                  'min_cw': (1, 10), 
                                  'subsample': (0.5, 1), 
                                  'colsample': (0.5, 1)
                                 })
```

```python
import time
start_time = time.time()
xgboostBO.maximize(init_points=10, n_iter = 50)
print('-'*53)
print('Final Results')
print('XGBOOST: %f' % xgboostBO.max['target'])
print("--- %s seconds ---" % (time.time() - start_time))
```
